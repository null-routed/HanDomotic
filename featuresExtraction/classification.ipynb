### Classificazione (SVM)

```python
import pandas as pd

# Carico i dati da file CSV
data = pd.read_csv("svm_dataset.csv")
data
```

### Queste due colonne causano overfitting, quindi le opzioni sono:
- si popola il DataFrame con più record
- si toglie queste due colonne

```python
data = data.drop(columns=["GradientDirection", "GradientNormSteep"])
```

### Assumo che il test set sia il 36%, altrimenti se gli si dà di più smatta

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=["class"])  # Features
y = data["class"]  # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.36, random_state=42)
```

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel="linear", probability=True)

svm_classifier.fit(X_train, y_train)
```

```python
# Valuta il modello sul set di test
accuracy = svm_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
prediction = svm_classifier.predict(X_test)
print("Predicted gesture: ", prediction)
```
